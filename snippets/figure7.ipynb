```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np 
import sys
import os
```

```python
file_path = '/home/erikz/sonnhammer/network_generation_algo/results/tables/2021-11-07_stability_analysis_itter10_datasets10_4method_4sizes.csv'
```

```python
plt.rcParams.update({'font.size': 26})

df = pd.read_csv(file_path,sep=",")
pd.to_numeric(df["min jtlide_lambda"])

df["log lambda"] = np.log2(abs(df["min jtlide_lambda"].values))
np.unique(df["Method"])
```

```python
fig, ax = plt.subplots(figsize=(16,12))

sns.lineplot(x="size", y="log lambda", markers=True, dashes=False,hue="Method",data=df,\
             hue_order = ['DAG', 'FFL', 'gnw', 'nx', 'randG'], ci=95, linewidth = 3,\
            palette = ['#fb659e','#00876c', '#e79053','#7dc9e1','#84b270'])
    
#ax.legend().set_title('Network')

plt.ylabel(r'$\log2$''($\lambda_{M}$)')
plt.xlabel("network size")
plt.legend(loc=2)

#os.chdir('../results/figures/')
#plt.savefig("figure7.svg")

plt.show()
```
